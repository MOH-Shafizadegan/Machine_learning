## Basic plots based on dataframes

```python
# Load the taxi trip dataset from a csv file
train_data = pd.read_csv("train_data.csv")

# 1: Display pairwise relationships between variables in the train_data dataframe
sns.pairplot(train_data)
```

```python
# 2:
# Group the data by salary and sex, and calculate the average count
average_count = train_data.groupby(['Salary', 'sex']).size().unstack()

# Plot the bar plot
average_count.plot(kind='bar')
plt.xlabel('Salary')
plt.ylabel('Count')
plt.title('Average Count of Males and Females in Each Salary Group')
plt.show()
```

```python
# 3: Plot the countplot
sns.countplot(x='Salary', data=train_data)
plt.xlabel('Salary')
plt.ylabel('Count')
plt.title('Count of Each Salary Group')
plt.show()
```

```python
# 4:
# Group the data by workclass and salary, and calculate the count
salary_by_workclass = train_data.groupby(['workclass', 'Salary']).size().unstack()

# Plot the bar plot
salary_by_workclass.plot(kind='bar', stacked=True)
plt.xlabel('Workclass')
plt.ylabel('Count')
plt.title('Count of Each Salary Group in Each Workclass')
plt.legend(title='Salary')
plt.show()
```

```python
# 5:
# Calculate the correlation matrix
correlation_matrix = train_data.corr()

# Plot the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Between Variables')
plt.show()
```

## Autograd, learning process and PyTorch <br>
1: Define a function that computes y_estimated=Wx+b (where W,x,b are Tensors with appropriate dimensions)<br>
2: Initialize a $2\times 2$ Tensor W and a $2\times 1$ Tensor b. Also activate the autograd feature for them. (Hint: Do not initialize them as zero, or the following part will not change)<br>
3: Define a function named loss_fn, that computes the MSE, given y_estimated and y_real as Tensors<br>
4: Generate a sample size of 20 (x, y_real pairs) with torch.rand. Use loss.backward() on the samples to compute the gradients and implement a gradient descent algorithm with a learning rate of your choice. Execute the loop 5 times. Print the resulting W and b aftereach iteration. (Hint: Search the internet about zeroing gradients in PyTorch) Also, note that you may not reach a low loss value, as all x,y are iid random numbers.


```python
# Define a function that computes y_estimated = Wx + b
def y_estimated(W, x, b):
    return torch.matmul(W, x) + b

# Define a function that computes the mean squared error (MSE) given y_estimated and y_real
def loss_fn(y_estimated, y_real):
    return torch.mean((y_estimated - y_real) ** 2)

```

```python
# Initialize a 2x2 tensor W and a 2x1 tensor b with random values and enable autograd
W = torch.rand(2, 2, requires_grad=True)
b = torch.rand(2, 1, requires_grad=True)

# Generate a sample size of 20 (x, y_real) pairs with torch.rand
x = torch.rand(2, 20)
y_real = torch.rand(2, 20)
```

```python
# Print the initial W and b
print("Initial W:")
print(W)
print("Initial b:")
print(b)
```

```python
# Set the learning rate for gradient descent
lr = 0.5

# Execute the loop 5 times
for i in range(5):
    # Compute y_estimated and loss using the function
    y_est = y_estimated(W, x, b)
    loss = loss_fn(y_est, y_real)
    print(f"Loss at iteration {i}: {loss.item()}")
    
    # Compute the gradients using backward()
    loss.backward()
    
    # Update W and b using gradient descent
    with torch.no_grad():
        W -= lr * W.grad
        b -= lr * b.grad
    # Zero the gradients for the next iteration
    W.grad.zero_()
    b.grad.zero_()

# Print the resulting W and b after the loop
print("Final W:")
print(W)
print("Final b:")
print(b)
```
