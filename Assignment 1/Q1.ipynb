## Basic plotting with matplotlib <br>
Draw 1000 samples from a binomial distribution with N=300 and p=0.4. Plot the histogram and the estimated cdf on a single figure

```python
# Set the random seed for reproducibility
np.random.seed(42)

# Draw 1000 samples from a binomial distribution with N=300 and p=0.4
samples = np.random.binomial(300, 0.4, 1000)

# Plot the histogram of the samples with 20 bins
plt.figure()
plt.hist(samples, bins=20, density=True, histtype="step")

# Calculate the cumulative counts of the samples
cum_counts = np.cumsum(np.bincount(samples))

# Normalize the cumulative counts to get the estimated cdf
cdf = cum_counts / cum_counts[-1]

# Plot the estimated cdf of the samples
plt.figure()
plt.plot(np.arange(len(cdf)), cdf)

# Add labels, title and legend to the plot
plt.xlabel("Number of successes")
plt.ylabel("Probability")
plt.title("Binomial distribution with N=300 and p=0.4")

# Show the plot
plt.show()
```

## SVD for denoising images <br>
Task 1: Convert the colored images to grayscale and display them. Then, use an SVD decoposition and partial reconstruction and plot the PSNR metric for each number of components used. Also, display some reconstructed images for a few values of the number of used singular values. <br>

    1: Create two noisy versions of the image, with almost the same PSNR and display them. Use a gaussian noise with a well-chosen amplitude for the first noisy image, and use a salt and pepper noise for the second one.<br>
    2: For each noisy image, plot the PSNR vs component number in SVD decomposition. Again, display the reconstructed iamges for a few values of the number of used singular values. <br>
    3: Determine the component number that maximizes the PSNR in each case, and explain why. Display the images for these component numbers as well. <br>
    4: Determine which of the two noise types can be removed more accurately with this method.
    

```python
# Load the colored images from a list of URLs
image1_gray = cv2.imread('city_pic.jpg', cv2.IMREAD_GRAYSCALE)
image2_gray = cv2.imread('Gorilla.jpg', cv2.IMREAD_GRAYSCALE)
gray_images = [image1_gray, image2_gray]

# display images
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
for i, gray_image in enumerate(gray_images):
  axes[i].imshow(gray_image, cmap="gray")
  axes[i].set_title(f"Image {i+1}")
plt.show()
```

```python
# Define a function to perform SVD decomposition and partial reconstruction of an image
def svd_reconstruct(image, k):
  # Perform SVD decomposition of the image matrix
  U, s, Vt = np.linalg.svd(image)
  # Select the first k singular values and vectors
  s_k = s[:k]
  U_k = U[:, :k]
  Vt_k = Vt[:k, :]
  # Reconstruct the image matrix from the reduced SVD
  image_k = U_k @ np.diag(s_k) @ Vt_k
  return image_k

# Define a function that calculates PSNR
def psnr (img1, img2):
  # Calculate the mean squared error between the images
  mse = np.mean ((img1 - img2) ** 2)
  # If MSE is zero, the images are identical
  if mse == 0:
    return math.inf
  # Otherwise, calculate and return the PSNR value
  max_pixel = 255.0 # Assuming 8-bit per channel images
  return -10 * math.log10 (mse / max_pixel ** 2)
```

```python
# Define a list of values for the number of singular values to use
k_values = [5, 10, 15, 20, 30, 50]

# Initialize an empty list to store the PSNR values for each image and k value
psnr_values = []

# Loop over each image
for i, gray_image in enumerate(gray_images):
  # Initialize an empty list to store the PSNR values for each k value
  psnr_values_i = []
  # Loop over each k value
  for k in k_values:
    # Reconstruct the image using k singular values
    image_k = svd_reconstruct(gray_image, k)
    # Calculate the PSNR between the original and reconstructed image
    psnr_k = psnr(gray_image, image_k)
    # Append the PSNR value to the list
    psnr_values_i.append(psnr_k)
  # Append the list of PSNR values to the main list
  psnr_values.append(psnr_values_i)

# Plot the PSNR metric for each image and k value
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
for i, psnr_values_i in enumerate(psnr_values):
  axes[i].plot(k_values, psnr_values_i)
  axes[i].set_xlabel("Number of singular values")
  axes[i].set_ylabel("PSNR (dB)")
  axes[i].set_title(f"Image {i+1}")
plt.show()

# Display some reconstructed images for a few values of k
fig, axes = plt.subplots(2, 4, figsize=(16, 8))
for i, gray_image in enumerate(gray_images):
  for j, k in enumerate([5, 10, 20, 50]):
    image_k = svd_reconstruct(gray_image, k)
    axes[i][j].imshow(image_k, cmap="gray")
    axes[i][j].set_title(f"Image {i+1} with k={k}")
plt.show()
```

```python
# Create a list for storing noisy images
noisy_images = []

# Loop over each image
for i, gray_image in enumerate(gray_images):

  # Create two noisy versions of the image with almost the same PSNR
  # Use a gaussian noise
  sigma = np.std(gray_image) / 150 # Choose a sigma value that gives a reasonable PSNR
  noisy_image_1 = util.random_noise(gray_image, mode="gaussian", mean=0, var=sigma**2)
  
  # Use a salt and pepper noise for the second noisy image
  noisy_image_2 = util.random_noise(gray_image, mode="s&p", amount=0.05)

  noisy_images.append(noisy_image_1)
  noisy_images.append(noisy_image_2)
  
  # Display the noisy images
  fig, axes = plt.subplots(1, 2, figsize=(8, 4))
  axes[0].imshow(noisy_image_1, cmap="gray")
  axes[0].set_title(f"Image {i+1} with gaussian noise")
  axes[1].imshow(noisy_image_2, cmap="gray")
  axes[1].set_title(f"Image {i+1} with salt and pepper noise")
  plt.show()
  
```

```python
# Initialize an empty list to store the PSNR values for each image and noise type and k value
psnr_values = []
max_k_for_psnr = []

for i, noisy_image in enumerate(noisy_images):
    
  # Initialize an empty list to store the PSNR values for each k value
  psnr_values_i_j = []
  # Loop over each k value
  for k in k_values:
    # Reconstruct the image using k singular values
    image_k = svd_reconstruct(noisy_image, k)
    # Calculate the PSNR between the original and reconstructed image

    original_image = gray_images[0] if i <= 1 else gray_images[1]

    psnr_k = psnr(original_image, image_k)

    psnr_values_i_j.append(psnr_k)

  # Determine the component number that maximizes the PSNR in each case
  max_psnr = max(psnr_values_i_j) # Find the maximum PSNR value
  max_k = k_values[psnr_values_i_j.index(max_psnr)] # Find the corresponding k value
  max_k_for_psnr.append(max_k)
  print(f"For image {i+1}, the component number that maximizes the PSNR is {max_k} with a PSNR of {max_psnr:.2f} dB")

  psnr_values.append(psnr_values_i_j)
```

```python
# Display some reconstructed images for a few values of k
fig, axes = plt.subplots(2, 4, figsize=(16, 8))
for i, noisy_image in enumerate(noisy_images[0:2]):
  for j, k in enumerate([5, 10, 20, 50]):
    image_k = svd_reconstruct(noisy_image, k)
    axes[i][j].imshow(image_k, cmap="gray")
    axes[i][j].set_title(f"Noisy image {i+1} with k={k}")
plt.show()

fig, axes = plt.subplots(2, 4, figsize=(16, 8))
for i, noisy_image in enumerate(noisy_images[2:4]):
  for j, k in enumerate([5, 10, 20, 50]):
    image_k = svd_reconstruct(noisy_image, k)
    axes[i][j].imshow(image_k, cmap="gray")
    axes[i][j].set_title(f"Noisy image {i+1} with k={k}")
plt.show()
```

## PCA for dimensionality reduction <br>
Load the iris dataset using the scikit-learn datasets submodule. Reduce the dimenssions of the dataset to two by using PCA, and visualize them. Each class must have a different color in the scatterplot. <br>
Explain PCA very briefly and how your code works as well.

<font color="cyan">
PCA stands for **principal component analysis**, which is a technique for reducing the dimensionality of a dataset by projecting it onto a lower-dimensional subspace that preserves most of the variance and information of the original data. PCA works by finding the directions (called principal components) that have the highest variance in the data, and transforming the data into a new coordinate system along these directions.
</font>

```python
# Load the iris dataset
iris = load_iris()

# Create a PCA object with 2 components
pca = PCA(n_components=2)

# Fit the PCA object to the iris data
pca.fit(iris.data)

# Transform the iris data into the new coordinate system
iris_pca = pca.transform(iris.data)

# Plot the transformed data
plt.scatter(iris_pca[:, 0], iris_pca[:, 1], c=iris.target, cmap='viridis')
plt.xlabel('First component')
plt.ylabel('Second component')
plt.title('PCA of iris dataset')
plt.show()
```

<font color="cyan">
As you can see, the plot shows that the three classes of iris flowers are well separated by the first two principal components, which capture most of the variation in the data. This means that PCA can be useful for dimensionality reduction, visualization, and clustering of high-dimensional data .
</font>
